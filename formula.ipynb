```python
# i
import matplotlib.pyplot as plt
import numpy as np

t = np.arange(0, 500, 0.01)
k= 0.02
V01= 50
V02=600

V1 = V01 * np.exp(-k*t) 
V2 = V02 * np.exp(-k*t) 

 
plt.plot(t, V1, color='r', label='V0=50')
plt.plot(t, V2, color='g', label='V0=600')


plt.title("i")
plt.xlabel("Values of t")
plt.ylabel("Values of V")

plt.legend()

plt.show()
```

```python
# ii
import matplotlib.pyplot as plt
import numpy as np

t = np.arange(0, 500, 0.01)
k= 0.02
V01= 50
V02=600

x1 = (V01/k) * (1 - np.exp(-k*t)) 
x2 = (V02/k) * (1 - np.exp(-k*t))  

 
plt.plot(t, x1, color='r', label='V0=50')
plt.plot(t, x2, color='g', label='V0=600')


plt.title("ii")
plt.xlabel("Values of t")
plt.ylabel("Values of x")

plt.legend()

plt.show()
```

```python
# iii
import matplotlib.pyplot as plt
import numpy as np

t = np.arange(0, 500, 0.01)
k= 0.02
V01= 50
V02=600
g= 9.8

V1 = -(g/k) + ((k*V01 + g)/ k)*(np.exp(-k*t)) 
V2 = -(g/k) + ((k*V02 + g)/ k)*(np.exp(-k*t)) 

 
plt.plot(t, V1, color='r', label='V0=50')
plt.plot(t, V2, color='g', label='V0=600')


plt.title("iii")
plt.xlabel("Values of t")
plt.ylabel("Values of V")

plt.legend()

plt.show()
```
